# Define GC Content Function

“If I give you a DNA sequence, how would you calculate GC content manually?”

```
def calculate_gc(sequence):
    sequence = sequence.upper()
    g = sequence.count("G")
    c = sequence.count("C")
    return (g + c) / len(sequence) * 100
```

“What does .count() do?”
“Why are we converting to uppercase?”

“What is the output?”

#  Extract Features

“What do you think this loop will produce?”
“Are we still working with sequences—or something else?”

```
data = []

for seq_record in sequences:
    seq = str(seq_record.seq)

    length = len(seq)
    gc_content = calculate_gc(seq)

    data.append({ #creating dictionary - question 3
        "Sequence_ID": seq_record.id,
        "Length": length,
        "GC_Content (%)": round(gc_content, 2)
    })

df = pd.DataFrame(data)
df
```

# Download Real FASTA from NCBI

```
import requests

# NCBI efetch URL for insulin gene (INS)
url = "https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi"

params = {
    "db": "nuccore",
    "id": "NM_000207",   # Human INS mRNA
    "rettype": "fasta",
    "retmode": "text"
}

response = requests.get(url, params=params)

# Check if request worked
if response.status_code == 200 and response.text.startswith(">"):
    with open("insulin.fasta", "w") as f:
        f.write(response.text)
    print("FASTA downloaded successfully!")
else:
    print("Error downloading FASTA. Check response.")
```

# View File

```
with open("insulin.fasta", "r") as f:
    print(f.read()[:500])  # show only first 500 chars
```

# Parse This Real Data

```
from Bio import SeqIO

real_sequences = list(SeqIO.parse("insulin.fasta", "fasta"))

print(f"Total sequences: {len(real_sequences)}")
```

# Inspect

```
for seq_record in real_sequences:
    print("ID:", seq_record.id)
    print("Length:", len(seq_record.seq))
    print("First 100 bases:", seq_record.seq[:100])
```

# Feature Extraction (Reuse Function)

```
real_data = []

for seq_record in real_sequences:
    seq = str(seq_record.seq)

    length = len(seq)
    gc_content = calculate_gc(seq)

    real_data.append({
        "Sequence_ID": seq_record.id,
        "Length": length,
        "GC_Content (%)": round(gc_content, 2)
    })

real_df = pd.DataFrame(real_data)
real_df
```
